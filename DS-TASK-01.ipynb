```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('polulation.csv',  encoding='utf-8')
df.head()
```

```python
# Print column names to debug
print(df.columns)

# Attempt to drop columns, ignoring errors if columns are not found
df_final = df.drop(['Country Code', 'Indicator Name', 'Indicator Code'], axis=1, errors='ignore')
```

```python
import matplotlib.pyplot as plt

# Inspect the DataFrame
print("Columns in df_final:", df_final.columns)
print(df_final.head())

# Strip any leading/trailing spaces from column names
df_final.columns = df_final.columns.str.strip()

# Check if 'Country Name' column exists and filter data
chosen_country = 'United States'

if 'Country Name' in df_final.columns:
    country_data = df_final[df_final['Country Name'] == chosen_country]

    # Check if country_data was successfully filtered
    if not country_data.empty:
        # Extract years and population data
        years = country_data.columns[2:].astype(int)
        population = country_data.iloc[:, 2:].values.flatten()

        # Plotting
        plt.figure(figsize=(10, 5))
        plt.bar(years, population, color='skyblue')
        plt.xlabel('Year')
        plt.ylabel('Population')
        plt.title(f'Population of {chosen_country} Over the Years')
        plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
        plt.tight_layout()       # Adjust layout to fit labels
        plt.show()
    else:
        print(f"No data available for {chosen_country}.")
else:
    print("Column 'Country Name' does not exist in df_final.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Example DataFrame with continuous data
data = {
    'Country': ['Country A', 'Country B', 'Country C', 'Country D', 'Country E'],
    'Population': [500000, 750000, 600000, 800000, 650000]
}

df = pd.DataFrame(data)

# Extract the population data
population = df['Population']

# Plotting a histogram
plt.figure(figsize=(8, 5))
plt.hist(population, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.title('Distribution of Population')
plt.show()
```
